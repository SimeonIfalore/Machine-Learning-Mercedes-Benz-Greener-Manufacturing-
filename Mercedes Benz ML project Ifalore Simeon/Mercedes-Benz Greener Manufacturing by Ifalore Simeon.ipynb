#### Performing  dimensionality reduction (PCA)

```python
# seperating y from the train dataFrame before scaling the features
X_train = train_df.iloc[:,1:]
y_train = train_df.iloc[:,0]

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled.shape
```

```python
y_train
```

```python
# applying the PCA model 
from sklearn.decomposition import PCA
```

```python
# PCA with 95% of x_train
sklearn_pca = PCA(n_components=0.95, random_state = 420)
```

```python
pca_x_train_transformed = sklearn_pca.fit_transform(X_train_scaled)
```

```python
print(pca_x_train_transformed.shape)
```

#### Dimensionality reduction for Test DataFrame. 

```python
# Scaling the data

scaling = MinMaxScaler()
test_scaled = scaling.fit_transform(test_df)
```

```python
test_scaled.shape
```

```python
# Performing Dimensionality reduction with PCA with 95% for test_df

pca_x_test_transformed = sklearn_pca.fit_transform(test_scaled)
```

```python

print(pca_x_test_transformed.shape) 
```

```python
pca_x_test_transformed.shape, pca_x_train_transformed.shape
```

#### Performing XGBOOST

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error as MSE

# fitting the model

model = xgb.XGBRegressor(objective="reg:linear",learning_rate=0.1)
model.fit(pca_x_train_transformed, y_train) 
y_pred = model.predict(pca_x_test_transformed)
y_pred
```

```python
y_pred.shape, pca_x_test_transformed.shape
```

```python

rmse = np.sqrt(MSE(y_train, y_pred))
print("RMSE : %f" %(rmse))
```
